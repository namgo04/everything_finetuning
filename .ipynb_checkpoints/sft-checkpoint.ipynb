```python
# Extract (instruction, chosen_response) from openbmb/UltraFeedback with a quality threshold.

from datasets import load_dataset

SPLIT = "train[:20000]"   # adjust as needed
MIN_SCORE = 5.0           # rows whose best completion scores below this are skipped

ds = load_dataset("openbmb/UltraFeedback", split=SPLIT)

def _num(x, default=None):
    try: return float(x)
    except: return default

def _as_int(x):
    try: return int(x)
    except: return None

def _resp(comp):
    # Some dumps use 'response', others 'content' or 'text'
    return (comp.get("response") or comp.get("content") or comp.get("text") or "").strip()

def _ratings_tuple(comp):
    ann = comp.get("annotations", {}) or {}
    h = _num(ann.get("helpfulness", {}).get("Rating"), 0.0)
    i = _num(ann.get("instruction_following", {}).get("Rating"), 0.0)
    t = _num(ann.get("truthfulness", {}).get("Rating"), 0.0)
    return h, i, t

def _score_tuple(comp):
    # Primary key: overall_score (float). Fallback to sum of ratings.
    os = _num(comp.get("overall_score"))
    if os is None:
        h, i, t = _ratings_tuple(comp)
        os = h + i + t
    # Return rich tuple for lexicographic max
    h, i, t = _ratings_tuple(comp)
    return (os, h, i, t)

def pick_preferred(example):
    comps = example.get("completions") or []
    if not comps: 
        return None

    # 1) prefer explicit correct_answers index if present and valid
    ca = example.get("correct_answers") or []
    idx = _as_int(ca[0]) if ca else None
    if idx is not None and 0 <= idx < len(comps):
        best_comp = comps[idx]
        best_score = _score_tuple(best_comp)[0]
    else:
        # 2) fallback: best by score tuple
        j = max(range(len(comps)), key=lambda k: _score_tuple(comps[k]))
        best_comp = comps[j]
        best_score = _score_tuple(best_comp)[0]

    text = _resp(best_comp)
    if not text: 
        return None
    if best_score is not None and best_score < MIN_SCORE:
        return None  # 3) quality filter

    return {
        "prompt": example.get("instruction", "").strip(),
        "chosen_text": text
    }

# Build extracted pairs
records = []
for ex in ds:
    rec = pick_preferred(ex)
    if rec: 
        records.append(rec)

records = records[:2000]

prompts = [r["prompt"] for r in records]
chosen_texts = [r["chosen_text"] for r in records]

print(f"Kept {len(chosen_texts)} examples (from {len(ds)} rows) with MIN_SCORE={MIN_SCORE}.")
print("Sample prompt:", (prompts[0][:160].replace("\n"," ") + "...") if prompts else "â€”")
print("Sample chosen:", (chosen_texts[0][:160].replace("\n"," ") + "...") if chosen_texts else "â€”")
```

```python
# --- Cluster chosen_texts with HF embeddings + KMeans, then preview 5 per cluster ---

!pip install -q sentence-transformers scikit-learn

from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans

# Params
K = 20                   # try 20 first; later try 30/40
BATCH_SIZE = 64

# 1) Embed
embedder = SentenceTransformer("all-mpnet-base-v2")
embs = embedder.encode(chosen_texts, batch_size=BATCH_SIZE, convert_to_numpy=True, show_progress_bar=True)

# 2) Cluster
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
labels = kmeans.fit_predict(embs)

# 3) Preview 5 examples per cluster
for cid in range(K):
    print(f"\n=== Cluster {cid} (showing up to 5) ===")
    shown = 0
    for i, lab in enumerate(labels):
        if lab == cid:
            print("â€”", chosen_texts[i][:200].replace("\n"," "), "...")
            shown += 1
            if shown == 5:
                break
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
embs_2d = tsne.fit_transform(embs)

plt.figure(figsize=(8,6))
scatter = plt.scatter(embs_2d[:,0], embs_2d[:,1], c=labels, cmap="tab20", s=10, alpha=0.6)
plt.title("Clusters of UF Responses (t-SNE 2D)")
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
plt.colorbar(scatter, label="Cluster ID")
plt.show()
```

```python
from datasets import load_dataset

SPLIT = "train[:2000]"   # slice for speed; remove the [:20000] for full
MIN_SCORE = 5.0           # drop weak answers

ds = load_dataset("openbmb/UltraFeedback", split=SPLIT)

def _num(x, default=None):
    try: return float(x)
    except: return default

def _as_int(x):
    try: return int(x)
    except: return None

def _resp(comp):
    return (comp.get("response") or comp.get("content") or comp.get("text") or "").strip()

def _ratings_tuple(comp):
    ann = comp.get("annotations", {}) or {}
    h = _num(ann.get("helpfulness", {}).get("Rating"), 0.0)
    i = _num(ann.get("instruction_following", {}).get("Rating"), 0.0)
    t = _num(ann.get("truthfulness", {}).get("Rating"), 0.0)
    return h, i, t

def _score_tuple(comp):
    os = _num(comp.get("overall_score"))
    if os is None:
        h, i, t = _ratings_tuple(comp)
        os = h + i + t
    h, i, t = _ratings_tuple(comp)
    return (os, h, i, t)

def pick_preferred(example):
    comps = example.get("completions") or []
    if not comps: return None

    # 1. Use correct_answers index if valid
    ca = example.get("correct_answers") or []
    idx = _as_int(ca[0]) if ca else None
    if idx is not None and 0 <= idx < len(comps):
        best = comps[idx]
    else:
        # 2. Else pick highest-scoring completion
        j = max(range(len(comps)), key=lambda k: _score_tuple(comps[k]))
        best = comps[j]

    score = _score_tuple(best)[0]
    text = _resp(best)
    if not text or (score is not None and score < MIN_SCORE):
        return None

    return {"prompt": example.get("instruction", "").strip(), "chosen": text}

# build cleaned records
records = []
for ex in ds:
    rec = pick_preferred(ex)
    if rec: records.append(rec)

print(f"Kept {len(records)} clean examples")
print(records[0])
```

```python
# we will the use perplexity metric here -- use the baseline model and check the probability the baseline model predicts the next token of the UF (golden_dataset) model. 
import torch, math, json
from transformers import AutoTokenizer, AutoModelForCausalLM

# load your eval set
eval_records = [json.loads(l) for l in open("eval.jsonl")]

# pick the base model (no FT yet!)
base_model = "meta-llama/Meta-Llama-3.1-8B-Instruct"

tok = AutoTokenizer.from_pretrained(base_model)
tok.pad_token = tok.eos_token
model = AutoModelForCausalLM.from_pretrained(
    base_model, device_map="auto", torch_dtype=torch.bfloat16
)
model.eval()

def ppl_on_eval(records, max_samples=100):
    total_loss, n = 0, 0
    for ex in records[:max_samples]:
        # pack prompt + reference answer in same format used for SFT
        text = f"<s>User:\n{ex['prompt']}\n\nAssistant:\n{ex['chosen']}</s>"
        enc = tok(text, return_tensors="pt", truncation=True, max_length=2048).to(model.device)
        with torch.no_grad():
            out = model(**enc, labels=enc["input_ids"])
        total_loss += out.loss.item()
        n += 1
    avg_loss = total_loss / n
    return math.exp(avg_loss)  # perplexity

ppl_base = ppl_on_eval(eval_records, max_samples=100)
print("Baseline perplexity:", ppl_base)
```

```python
# here we are generating outputs from the baseline model using the a100. only using an 8b parameter model. we will get the agent answers from the
# eval-set since that is what we will use to benchmark
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch, json

base_model = "meta-llama/Meta-Llama-3.1-8B-Instruct"
tok = AutoTokenizer.from_pretrained(base_model)
tok.pad_token = tok.eos_token
model = AutoModelForCausalLM.from_pretrained(
    base_model, device_map="auto", torch_dtype=torch.bfloat16
)
model.eval()

eval_records = [json.loads(l) for l in open("eval.jsonl")]

def generate_base_outputs(records, max_samples=1000):
    outs = []
    for ex in records[:max_samples]:
        prompt = f"<s>User:\n{ex['prompt']}\n\nAssistant:\n"
        inputs = tok(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            out = model.generate(
                **inputs,
                max_new_tokens=256,
                temperature=0.2,
                top_p=0.9
            )
        ans = tok.decode(out[0], skip_special_tokens=True).split("Assistant:\n")[-1].strip()
        outs.append({"prompt": ex["prompt"], "base_response": ans, "reference": ex["chosen"]})
    return outs

base_outputs = generate_base_outputs(eval_records, max_samples=1000)

with open("base_outputs.jsonl", "w") as f:
    for r in base_outputs:
        f.write(json.dumps(r) + "\n")

print("Saved", len(base_outputs), "base outputs")
```
